```
for id_sign in dict_signs.keys():
  given_signs = X_train[y_train == id_sign]
  plt.figure(figsize=(15,5))
  for i in range(9):
    plt.subplot('19{0}'.format(i+1))
    plt.imshow(given_signs[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```
cnt = Counter(y_train).most_common()
id_labes, cnt_labels = zip(*cnt)
ids = range(len(id_labes))

plt.figure(figsize=(15,5))
plt.bar(ids, cnt_labels)

plt.xlabel('Znaki')
#id trzeba odkrecic
#id_labels[10]
#dict_signs[7]
#dict_signs[id_labes[10]]
labels = [dict_signs[id_labes[id_]] for id_ in id_labes]
plt.xticks(ids, labels, rotation = 'vertical')
plt.title('Znaki drogowe wg.wystapien')
plt.show()
```
